# Imputasi Missing Value

```python
Y = ['type']
X_int = ['age','fnlwgt','education-num','capital-gain','capital-loss','hours-per-week']
X_cat = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'native-country']
```

```python
data_Y = data[Y]
data_int = data[X_int].astype(float)
data_cat = data[X_cat]
```

```python
print(data_Y['type'].unique())
```

```python
data_cat_array = []
for class_name in data_Y['type'].unique():
    data_cat_1 = data_cat[data_Y['type']==class_name]
    data_cat_clean_1 = data_cat_1.where(pd.notnull(data_cat_1), data_cat_1.mode(), axis='columns')
    data_cat_array.append(data_cat_clean_1)
```

```python
data_cat_clean = pd.concat(data_cat_array)
```

```python
data_num_w_class = pd.concat([data_int, data_Y], axis=1)
data_done_missing = pd.concat([data_num_w_class, data_cat_clean], axis=1)
data_done_missing.head(n=20)
```

```python
data_done_missing.isnull().sum()
```

# Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data2 = data_done_missing
```

```python
data2.head(n=20)
```

```python
#Standardize untuk data numerik = 6 variabel
data2['fnlwgt'] = StandardScaler().fit(data2[['fnlwgt']]).transform(data2[['fnlwgt']])
data2['education-num'] = StandardScaler().fit(data2[['education-num']]).transform(data2[['education-num']])
data2['capital-gain'] = StandardScaler().fit(data2[['capital-gain']]).transform(data2[['capital-gain']])
data2['capital-loss'] = StandardScaler().fit(data2[['capital-loss']]).transform(data2[['capital-loss']])
data2['hours-per-week'] = StandardScaler().fit(data2[['hours-per-week']]).transform(data2[['hours-per-week']])
data2
```

# Outlier

```python
X_int = ['age','fnlwgt','education-num','capital-gain','capital-loss','hours-per-week']
data2_int = data2[X_int].astype(float)
data_num_w_class2 = pd.concat([data2_int, data_Y], axis=1)
```
