```python
#MULTIPLE KEY
buah = pd.DataFrame ({'Kode Barang' : ['C01', 'E02', 'C03', 'A04', 'D05', 'E06', 'B07', 'A08', 'A09', 'B10'],
                      'Stock' : ['Penuh', 'Ada', 'Habis', 'Penuh', 'Penuh', 'Ada', 'Habis', 'Penuh', 'Ada', 'Habis'],
                      'Nama Buah' : ['Pear', 'Manggis', 'Alpukat', 'Belimbing', 'Lemon', 'Kelengkeng', 'Rambutan', 'Pepaya', 'Nangka', 'Nanas'],
                      'Jumlah Pemesanan (kg)' : ['3', '4', '7', '2', '3', '5', '6', '2', '4', '8']})
supplier = pd.DataFrame({'Kode Barang' : ['C01', 'E02', 'C03', 'A04', 'D05', 'E06', 'B07', 'A08', 'A09', 'B10'],
                         'Stock' : ['Ada', 'Ada', 'Ada', 'Habis', 'Penuh', 'Penuh', 'Habis', 'Habis', 'Ada', 'Habis'],
                         'Harga per kg (rupiah)' : ['26.900', '12.900', '35.000', '15.000', '37.500', '35.000', '12.000', '8.000', '9.000', '7.500'],
                         'Supplier' : ['C', 'E', 'C', 'A', 'D', 'E', 'B', 'A', 'A', 'B']})
data4 = pd.merge(buah,supplier, on=['Kode Barang', 'Stock'])
data4
```

```python
#MERGE METHODS 
data_left = pd.merge(buah,supplier, how='left', on=['Kode Barang', 'Stock'])
data_left #(berdasarkan data frame buah)
```

```python
#MERGE METHODS 
data_right = pd.merge(buah,supplier, how='right', on=['Kode Barang', 'Stock'])
data_right #(berdasarkan data frame supplier)
```

```python
data_outer = pd.merge(buah, supplier, how='outer', on=['Kode Barang', 'Stock']) #semua data dimasukkin
data_outer
```

```python
data_outer = pd.merge(buah, supplier, how='inner', on=['Kode Barang', 'Stock']) #yang cocok aja yang keluar di output
data_outer
```

```python
#JOINING
buah = pd.DataFrame ({'Nama Buah' : ['Pear', 'Manggis', 'Alpukat', 'Belimbing', 'Lemon', 'Kelengkeng', 'Rambutan', 'Pepaya', 'Nangka', 'Nanas'],
                     'Jumlah Pemesanan (kg)' : ['3', '4', '7', '2', '3', '5', '6', '2', '4', '8']},
                      index =['A01', 'A12', 'A03', 'A04', 'A05', 'A16', 'A07', 'A08', 'A09', 'A10'])
supplier = pd.DataFrame({'Harga per kg (rupiah)' : ['26.900', '12.900', '35.000', '15.000', '37.500', '35.000', '12.000', '8.000', '9.000', '7.500'],
                        'Supplier' : ['C', 'E', 'C', 'A', 'D', 'E', 'B', 'A', 'A', 'B']},
                        index =['A01', 'A02', 'A03', 'A14', 'A15', 'A06', 'A07', 'A08', 'A09', 'A10'])
data5 = buah.join(supplier) #data buah ditambahkan data supplier
data5
```

```python
data6 = supplier.join(buah, how='outer') #seluruh data digabung
data6
```

```python
data7 = supplier.join(buah, how='inner') #Hanya data yang lengkap/cocok
data7
```

```python
#CONCATENATING
D1 = pd.DataFrame({'Kode Barang':['A01', 'A02', 'A03'],
                   'Nama Buah' : ['Pear', 'Manggis', 'Alpukat'],
                   'Harga per kg (rupiah)' : ['26.900', '12.900', '35.000'],
                   'Supplier' : ['C', 'E', 'C']})
D2 = pd.DataFrame({'Kode Barang':['A04', 'A05', 'A06'],
                   'Nama Buah' : ['Belimbing', 'Lemon', 'Kelengkeng'],
                   'Harga per kg (rupiah)' : ['15.000', '37.500', '35.000'],
                   'Supplier' : ['A', 'D', 'E']})
D3 = pd.DataFrame({'Kode Barang':['A07', 'A08', 'A09'],
                   'Nama Buah' : ['Rambutan', 'Pepaya', 'Nangka'],
                   'Harga per kg (rupiah)' : ['12.000', '8.000', '9.000'],
                   'Supplier' : ['B', 'A', 'A']})
frames = [D1, D2, D3]
data11 = pd.concat(frames)
data11
```
