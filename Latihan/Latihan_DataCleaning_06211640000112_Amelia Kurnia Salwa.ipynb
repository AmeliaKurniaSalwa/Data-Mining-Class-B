## Imputation Missing Value

```python
#Jumlah data missing value (NA)
np.sum(data.isnull())
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
#Imputasi missing value untuk data numerik
#data numerik di imputasi dengan nilai mean variabel
data['Bilirubin'] = data['Bilirubin'].fillna((data['Bilirubin'].mean())) 
data['Alk Phospat'] = data['Alk Phospat'].fillna((data['Alk Phospat'].mean()))
data['SGOT'] = data['SGOT'].fillna((data['SGOT'].mean()))
data['Albumin'] = data['Albumin'].fillna((data['Albumin'].mean()))
data['Protime'] = data['Protime'].fillna((data['Protime'].mean()))
data.isnull().any()
```

```python
#Imputasi missing value untuk data kategorik
#data kategorik di imputasi dengan nilai modus
data['Liver Big'] = data.fillna(data['Liver Big'].value_counts().index[0])
data['Liver Firm'] = data.fillna(data['Liver Firm'].value_counts().index[0])
data['Spleen Pabable'] = data.fillna(data['Spleen Pabable'].value_counts().index[0])
data['Spiders'] = data.fillna(data['Spiders'].value_counts().index[0])
data['Ascites'] = data.fillna(data['Ascites'].value_counts().index[0])
data['Varices'] = data.fillna(data['Varices'].value_counts().index[0])
data.isnull().any()
```

```python
#Imputasi missing value untuk data kategorik
#data kategorik di imputasi dengan nilai modus variabel
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Encode untuk data kategorik
data['Class'] = LabelEncoder().fit_transform(data['Class'])
data['Sex'] = LabelEncoder().fit_transform(data['Sex'])
data['Steroid'] = LabelEncoder().fit_transform(data['Steroid'])
data['Antivirals'] = LabelEncoder().fit_transform(data['Antivirals'])
data['Fatigue'] = LabelEncoder().fit_transform(data['Fatigue'])
data['Malaise'] = LabelEncoder().fit_transform(data['Malaise'])
data['Anaroxia'] = LabelEncoder().fit_transform(data['Anaroxia'])
data['Liver Big'] = LabelEncoder().fit_transform(data['Liver Big'])
data['Liver Firm'] = LabelEncoder().fit_transform(data['Liver Firm'])
data['Spleen Pabable'] = LabelEncoder().fit_transform(data['Spleen Pabable'])
data['Spiders'] = LabelEncoder().fit_transform(data['Spiders'])
data['Ascites'] = LabelEncoder().fit_transform(data['Ascites'])
data['Varices'] = LabelEncoder().fit_transform(data['Varices'])
data['Histology'] = LabelEncoder().fit_transform(data['Histology'])
data.head(n=10)
```

# Standardize
