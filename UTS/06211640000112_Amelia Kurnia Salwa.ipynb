# UTS DATA MINING NOMOR 5

```python
import numpy as np
import pandas as pd
from apyori import apriori 
```

```python
#Membaca data
data = pd.read_csv('online_retail.csv')
data.head(n=5)
```

```python
#Mendetksi missing value
np.sum(data.isnull())
```

## Terdapat missing value pada data online retail, maka ditambahkan kode barang daftar belanja

```python
for i, j in data[data['Description'].isnull()].iterrows():data['Description'][i]="ItemCode-"+str(j['StockCode'])
```

```python
np.sum(data.isnull())
```

## Menggabungkan barang-barang yang termasuk dalam 1 transaksi 

```python
#Mengelompokkan barang-barang dalam satu trnasaksi dengan bantuan data pada kolom Invoice No
X = data.groupby('InvoiceNo')
transaksi = []
for i, j in X:
    transaksi.append(list(j['Description'].map(str)))
```

```python
pd.DataFrame(transaksi)
```

## Perhitungan association rule

```python
#Syarat apriori
association_rules=apriori(transaksi, min_support=0.0055,
                         min_confidence=0.4,
                         min_lift=3,
                         min_length=2)
association_result=list(association_rules)
```

```python
association_result[10]
```

```python
for item in association_result:
#Indeks pertama dari inner list
#berisikan item yang dibeli serta item tambahan yang akan dibeli
    pair = item[0]
    items = [X for X in pair]
    print("Rule       = " + items [0] + " -> " + items [1])
    
    #Indeks kedua dari inner list
    print("Support   = " + str(item[1]))
    
    #Indeks ketiga dari daftar terletak di urutan ke-10
    #Indeks ketiga dari inner list
    
    print("Confidence = " + str(item[2][0][2]))
    print("Lift      = " + str(item[2][0][3]))
    
    print("______________________________________________________________")
```
